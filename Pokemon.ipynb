```python
import urllib.request
from urllib.request import urlopen
from bs4 import BeautifulSoup

codes = reversed(range(1588,6000))
for code in codes:
    print(code)
    url_text = "https://asia.pokemon-card.com/tw/card-search/detail/"+str(code).zfill(4)
    htmldata = urlopen(url_text)
    soup = BeautifulSoup(htmldata, 'html.parser')
    words = soup.find_all('h1')
    skills = soup.find_all("span", class_="skillName")
    skills = "_" + "_".join([skill.text for skill in skills ])
    if words[0].text.replace(" ","").split()[0]=="卡牌搜尋結果":#該code沒有卡片 
        pass
    else:
        if len(words[0].text.replace(" ","").split())==2: #代表是寶可夢卡
            stage = (words[0].text.replace(" ","").split()[0])
            name  = (words[0].text.replace(" ","").split()[1])
        else:
            non_montser = soup.find_all('h3') #代表是非寶可夢卡
            stage = non_montser[0].text.replace(" ","").split()[0]
            name  = (words[0].text.replace(" ","").split()[0])
            skill = None
        url_img = 'https://asia.pokemon-card.com/tw/card-img/tw0000'+ str(code).zfill(4) +'.png'
        urllib.request.urlretrieve(url_img
        ,"/Users/HaHaChang/Downloads/all/"+stage+"_"+name + skills+ ".png")
```
